```python
# Extracts values into a list

def extract_values(text):
    if pd.isna(text):
        return []
    else:
        return re.findall(r'[a-zA-Z]+\s+\d+', text)
```

```python
def list_to_dict(lst):
  """Converts a list of pairs to a dictionary."""
  return {item.split()[0]: int(item.split()[1]) for item in lst}

# my_list = ['b 1', 'lb 4', 'nb 1', 'w 3']
# my_dict = list_to_dict(my_list)
# print(my_dict)  # Output: {'b': 1, 'lb': 4, 'nb': 1, 'w': 3}
```

```python
# Extracting exrtas into new columns for team 1

df_scores['extracted_extras_team1'] = df_scores['Team_1_Extras_given'].apply(extract_values)
df_scores['extracted_extras_team1'] = df_scores['extracted_extras_team1'].apply(list_to_dict)
df_scores
```

```python

df_scores['T1_Wides_given'] = df_scores['extracted_extras_team1'].apply(lambda x : x['w'] if len(x)!=0 and 'w' in x else 0)
df_scores['T1_Byes_given'] = df_scores['extracted_extras_team1'].apply(lambda x : x['b'] if len(x)!=0 and 'b' in x else 0)
df_scores['T1_LBs_given'] = df_scores['extracted_extras_team1'].apply(lambda x : x['lb'] if len(x)!=0 and 'lb' in x else 0)
df_scores['T1_NBs_given'] = df_scores['extracted_extras_team1'].apply(lambda x : x['nb'] if len(x)!=0 and 'nb' in x else 0)
```

```python
# Extracting exrtas into new columns for team 2

df_scores['Team_2_Extras_given'] = df_scores['Team_2_Extras_given'].apply(extract_values)
df_scores['Team_2_Extras_given'] = df_scores['Team_2_Extras_given'].apply(list_to_dict)

df_scores['T2_Wides_given'] = df_scores['extracted_extras_team1'].apply(lambda x : x['w'] if len(x)!=0 and 'w' in x else 0)
df_scores['T2_Byes_given'] = df_scores['extracted_extras_team1'].apply(lambda x : x['b'] if len(x)!=0 and 'b' in x else 0)
df_scores['T2_LBs_given'] = df_scores['extracted_extras_team1'].apply(lambda x : x['lb'] if len(x)!=0 and 'lb' in x else 0)
df_scores['T2_NBs_given'] = df_scores['extracted_extras_team1'].apply(lambda x : x['nb'] if len(x)!=0 and 'nb' in x else 0)
```
